```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

The following dataset is from https://www.kaggle.com/datasets/anshtanwar/residential-ev-chargingfrom-apartment-buildings, which has data from this article https://www.sciencedirect.com/science/article/pii/S0378778821002073?via%3Dihub, which is EV data from a research project done in a suburb in Norway.

```python
dataset1 = pd.read_csv("data/Dataset 1_EV charging reports.csv", sep=";", decimal=",")
dataset1
```

The following is data from https://dataverse.harvard.edu/dataset.xhtml?persistentId=doi:10.7910/DVN/NFPQLW, which contains data from Harvard dataverse's repo of EV charging sessions across 105 stations at a workplace charging program.

```python
dataset2 = pd.read_csv("data/station_data_dataverse.csv")
dataset2
```

The following dataset is from https://data.nrel.gov/submissions/182, which tracks 300 vehicles over 4 year period from NREL.

```python
dataset3 = pd.read_csv("data/NRELs Charging Data - Raw.csv")
dataset3
```

THe following dataset is from https://www.kaggle.com/datasets/venkatsairo4899/ev-charging-station-usage-of-california-city/data, which is from https://data.cityofpaloalto.org/dataviews/257812/electric-vehicle-charging-station-usage-july-2011-dec-2020/. It is about a 9 year collection of EV charging station usage from Palo Alto.

```python
dataset4 = pd.read_csv("data/EVChargingStationUsage.csv")
dataset4
```

CONSIDER: only using Palo Alto Data since it has 250k+ entries, but it does not say how much kW the chargers were, so there can be an assumption it is a standard 3.6 kW or a 7.2 kW charger, multiplied by the hours it charged to obtain kWh (an energy unit).

```python
dataset1.drop(dataset1.columns.difference(['Start_plugin','End_plugout', 'El_kWh']), axis=1, inplace=True)
# print(dataset1)
dataset2.drop(dataset2.columns.difference(['created','ended', 'kwhTotal']), axis=1, inplace=True)
# print(dataset2)
dataset3.drop(dataset3.columns.difference(['start_charge','termin_charge', 'energy_charged']), axis=1, inplace=True)
print(dataset3)
dataset4.drop(dataset4.columns.difference(['Start Date','End Date', 'Charging Time (hh:mm:ss)']), axis=1, inplace=True)
print(dataset4)

dataset1.dropna(inplace=True)
dataset2.dropna(inplace=True)
dataset3.dropna(inplace=True)
dataset4.dropna(inplace=True)
```

```python
start_datetime1 = pd.to_datetime(dataset1['Start_plugin'], format="%d.%m.%Y %H:%M")
end_datetime1 = pd.to_datetime(dataset1['End_plugout'], format="%d.%m.%Y %H:%M")
dataset1["year_start"] = start_datetime1.dt.year
dataset1["month_start"] = start_datetime1.dt.month
dataset1["day_start"] = start_datetime1.dt.day
dataset1["hour_start"] = start_datetime1.dt.hour
dataset1["minute_start"] = start_datetime1.dt.minute
dataset1["year_end"] = end_datetime1.dt.year
dataset1["month_end"] = end_datetime1.dt.month
dataset1["day_end"] = end_datetime1.dt.day
dataset1["hour_end"] = end_datetime1.dt.hour
dataset1["minute_end"] = end_datetime1.dt.minute
dataset1["duration"] = end_datetime1 - start_datetime1
dataset1.drop(["Start_plugin", "End_plugout"], axis=1, inplace=True)
dataset1.rename(columns={"El_kWh": "kWh"}, inplace=True)
dataset1
```

```python
dataset2["created"] = "2" + dataset2["created"].str[1:]
dataset2["ended"] = "2" + dataset2["ended"].str[1:]
start_datetime2 = pd.to_datetime(dataset2['created'], format="%Y-%m-%d %H:%M:%S")
end_datetime2 = pd.to_datetime(dataset2['ended'], format="%Y-%m-%d %H:%M:%S")
dataset2["year_start"] = start_datetime2.dt.year
dataset2["month_start"] = start_datetime2.dt.month
dataset2["day_start"] = start_datetime2.dt.day
dataset2["hour_start"] = start_datetime2.dt.hour
dataset2["minute_start"] = start_datetime2.dt.minute
dataset2["year_end"] = end_datetime2.dt.year
dataset2["month_end"] = end_datetime2.dt.month
dataset2["day_end"] = end_datetime2.dt.day
dataset2["hour_end"] = end_datetime2.dt.hour
dataset2["minute_end"] = end_datetime2.dt.minute
dataset2["duration"] = end_datetime2 - start_datetime2
dataset2.drop(["created", "ended"], axis=1, inplace=True)
dataset2.rename(columns={"kwhTotal": "kWh"}, inplace=True)
dataset2
```

```python
start_datetime3 = pd.to_datetime(dataset3['start_charge'], format="%Y-%m-%d %H:%M:%S")
end_datetime3 = pd.to_datetime(dataset3['termin_charge'], format="%Y-%m-%d %H:%M:%S")
dataset3["year_start"] = start_datetime3.dt.year
dataset3["month_start"] = start_datetime3.dt.month
dataset3["day_start"] = start_datetime3.dt.day
dataset3["hour_start"] = start_datetime3.dt.hour
dataset3["minute_start"] = start_datetime3.dt.minute
dataset3["year_end"] = end_datetime3.dt.year
dataset3["month_end"] = end_datetime3.dt.month
dataset3["day_end"] = end_datetime3.dt.day
dataset3["hour_end"] = end_datetime3.dt.hour
dataset3["minute_end"] = end_datetime3.dt.minute
dataset3["duration"] = end_datetime3 - start_datetime3
dataset3.drop(["start_charge", "termin_charge"], axis=1, inplace=True)
dataset3.rename(columns={"energy_charged": "kWh"}, inplace=True)
dataset3
```

```python
start_datetime4 = pd.to_datetime(dataset4['Start Date'], format="%m/%d/%Y %H:%M", errors="coerce")
end_datetime4 = pd.to_datetime(dataset4['End Date'], format="%m/%d/%Y %H:%M", errors="coerce")
start_datetime1.dropna(inplace=True)
end_datetime1.dropna(inplace=True)
dataset4["year_start"] = start_datetime4.dt.year
dataset4["month_start"] = start_datetime4.dt.month
dataset4["day_start"] = start_datetime4.dt.day
dataset4["hour_start"] = start_datetime4.dt.hour
dataset4["minute_start"] = start_datetime4.dt.minute
dataset4["year_end"] = end_datetime4.dt.year
dataset4["month_end"] = end_datetime4.dt.month
dataset4["day_end"] = end_datetime4.dt.day
dataset4["hour_end"] = end_datetime4.dt.hour
dataset4["minute_end"] = end_datetime4.dt.minute
dataset4["duration"] = end_datetime4 - start_datetime4
duration = pd.to_datetime(dataset4["Charging Time (hh:mm:ss)"], format="%H:%M:%S")
dataset4["kWh"] = (duration.dt.hour + duration.dt.minute / 60 + duration.dt.second / 3600) * 7.2 # assumes standard Level 2 charger which is 7.2 kW
dataset4.drop(["Start Date", "End Date", "Charging Time (hh:mm:ss)"], axis=1, inplace=True)
dataset4.dropna(inplace=True)
dataset4
```

Below is the first three datasets concated together.

```python
datasets_without_assumptions = pd.concat([dataset1, dataset2, dataset3])
datasets_without_assumptions
```

Below is all four datasets concated together, which includes the assumption made for the last dataset.

```python
datasets_with_assumptions = pd.concat([dataset1, dataset2, dataset3, dataset4])
datasets_with_assumptions
```

Export the data from here:

```python
datasets_without_assumptions.to_csv("datasets_without_assumptions.csv")
datasets_with_assumptions.to_csv("datasets_with_assumptions.csv")
```
